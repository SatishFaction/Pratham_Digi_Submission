```python

import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=["Pratham"])
# Interest by Region
df = pytrend.interest_by_region()
df.head(20)
```

```python
df.reset_index().plot(x="geoName", y="Pratham", figsize=(120, 10), kind ="bar")
```

```python
# Get Google Hot Trends data
df = pytrend.trending_searches(pn="india")
df.head()
```

```python
# Get Google Keyword Suggestions
keywords = pytrend.suggestions(keyword='Pratham foundation')
df = pd.DataFrame(keywords)
df.drop(columns= 'mid')   # This column makes no sense
```
